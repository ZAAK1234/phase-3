```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE()
X,y=sm.fit_resample(X,y)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier,ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
model={'LogisticRegression':LogisticRegression(),'DecisionTreeClassifier':DecisionTreeClassifier(),
       'RandomForestClassifier':RandomForestClassifier(),'AdaBoostClassifier':AdaBoostClassifier(),
      'GradientBoostingClassifier':GradientBoostingClassifier(),'ExtraTreesClassifier':ExtraTreesClassifier(),
      'SVC':SVC(),'XGBClassifier':XGBClassifier(),'Neural_Network':MLPClassifier()}
model_accuracies={}
for k in range(50):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=k)
    for i,j in model.items():
        j.fit(X_train,y_train)
        pred=j.predict(X_test)
        acc=accuracy_score(y_test,pred)
        model_accuracies[j]=acc
        print(f'Model is {i} with Random State {k}')
        print(f'Accuracy_score is {acc}')
        print(confusion_matrix(y_test,pred))
        print(classification_report)
        print('\n')
    
```

```python
best_model = max(model_accuracies, key=model_accuracies.get)

print("Best model:", best_model)
print("Accuracy score:", model_accuracies[best_model])
```

### HyperParameter Tunning with XGBClassifier

```python
from sklearn.model_selection import RandomizedSearchCV
param_grid={
    'max_depth':[None,5,10],
    'n_estimators':[50,100,200],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}

et=ExtraTreesClassifier()
for i in range(4,10):
    rd=RandomizedSearchCV(et,param_grid,cv=i)
    rd.fit(X_train,y_train)
    best_param=rd.best_params_
    print(f'Best HyperParameter is {best_param}')
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=38)
et=ExtraTreesClassifier(n_estimators= 100, min_samples_split= 2, min_samples_leaf= 1, max_depth=30,random_state=72)
et.fit(X_train,y_train)
pred=et.predict(X_test)
acc=accuracy_score(y_test,pred)
con=confusion_matrix(y_test,pred)
print(f'Accuracy Score of ExtraTreesClassifier with Hyper Parameter Tunning is {acc}')
print(con)
print(classification_report)
```
